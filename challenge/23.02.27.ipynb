```python
# 'jd'열에 있는 기사를 토큰화 함수를 적용하여 'tokenized'열에 저장한다.
df['tokenized'] = df['jd'].apply(tokenizer.nouns)
```

```python
df['tokenized']
```

```python
# 불용어 리스트 만들기
stop_words = ['주요업무', '자격요건', '우대사항', '혜택', '및', '복지', '분']
```

```python
# 불용어 제거 코드
df['tokenized'] = df['tokenized'].apply(lambda x: [item for item in x if item not in stop_words and len(item) > 1])
```

```python
# 사용하고자 하는 폰트의 경로. Colab에서는 이 경로를 사용하시면 됩니다.
fontpath = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
```

```python
# 카테고리별 단어 빈도 수 파악하기
from collections import Counter

word_count = Counter(df['tokenized'])
print(word_count.most_common(20))
```

```python
# 사회 뉴스 기사의 워드클라우드
plt.figure(figsize = (15,15))
temp_data = ' '.join(df['tokenized'])
wc = WordCloud(max_words = 2000 , width = 1800 , height = 1100, font_path = fontpath).generate(df['tokenized'])
plt.imshow(wc, interpolation = 'bilinear')
```

```python
nouns = df['tokenized']  # 명사만 추출

words = [n for n in nouns if len(n) > 1] # 단어의 길이가 1개인 것은 제외
```
